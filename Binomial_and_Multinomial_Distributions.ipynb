##Binary variables and Binomial Distribution
Binary variables are random variables which take on two random values ie. x ∈ {0, 1}. Let us take example of flipping a coin. When we flip the coin we will randomly get heads or tails. The outcome x will be 1 for heads and 0 for tails.

###Visualizing Bernoulli distribution

image.png

In an ideal world the probability to appear heads or tails while flipping the coin will be 50% for each. That is each side of the coin has the same probability to occur. But we do not live in an ideal world but we live in a practical world. Here we are going to flip a damaged coin in which the probability of getting heads is not equal to the probability of getting tails.

We want to find the probability of getting heads of the damaged coin. So x = 1. 'µ' represents the probability of getting heads (x = 1). So the below line explains that the probability 'µ' of the damaged coin to land on heads (x = 1) is the probability 'µ' itself.

> p(x = 1|µ) = µ

Likewise the probability of coin to get tails will be:

> p(x = 0|µ) = 1 − µ

So this concept of tossing a coin comes under Bernoulli's distribution.

> $Bern(x|µ) = µ^x(1 − µ)^{(1-x)}$

There is nothing to fear about this equation. Let us first find the probability of getting heads. So x = 1. Putting the value of x in the equation we will get 'µ' probability as the result.

It is simple to understand because when we put x equal to 1, then (1 - 1) = 0, and anything raised to zero is 1.

> $µ ^ 1 = µ$

So, $Bern(x = 1|µ) = µ$

Let us first find the probability of getting tails. So x = 0. Putting the value of x in the equation we will get '1 - µ' probability as the result.

Again simple because when we put x equal to 0, then (1 - 0) = 1. So, it will be (1 - µ).

 Now $µ ^ 0 = 1$ because anything raised to zero is 1.

So, 
> $Bern(x|0) = (1 − µ)$

##Random Variables
A Random variable consists of values which come from random distribution. There are two types of random variables.
1. Discrete Random Variable (D.R.V) 
2. Continuous Random Variable (C.R.V)

## Discrete Random Variables
The discrete random variables takes finite number of possible values within a particular range. We can take an example of casting a dice. The number of possible outcomes is 6. So X = [1, 2, 3, 4, 5, 6]. So P[x = Xi] = Pi.

The Discrete Random Variables should have this properties- 
1. 0 < Pi < 1  : The probability of a DRV should range between zero (False) and one (True)
2. P1 + P2 + P3 + P4 +... Pn = 1
The sum of probabilities of total number of elements in DRV should be equal to one. For example - Dice. There are six faces of dice. X = [1, 2, 3, 4, 5, 6]. The probability of 1 appearing as the face (P1) + probability of 1 appearing as the face (P2) + probability of 1 appearing as the face (P3) ... probability of 6 appearing as the face (P6) there sum will be one.

##Continuous Random Variables
A continuous random variable consists of infinite values. Latest taken example of the height of students in a class. Here the height of the students is continuous in nature ie. 163.7, 150.4, 175.9 and so on. Discrete variable contains specific values like [0, 1, 2, 3]. But in this case we are getting continuous point values.

Since the continuous random variable is continuous in nature, its values are represented by area under the curve (or the integral). The probability of a continuous random variable that is continuous distribution functions are the functions that take on continuous values.

The probability of observing any single value in the distribution is equal to zero, since the number of values that are assumed by the random variable is infinite.

For example a random variable X will take all values within a interval of real numbers.

The CVR should satisfy the following conditions - 
1. The curve is always positive. $[p(x) > 0$ for all values of $x$]
2. The total area under the curve is equal to 1.

The probability of discrete variable will lie between 0 and 1 whereas the probability of the continuous random variable is always greater than 1. It could be very near to zero but it will be always positive.

The `uniform` function generates a uniform continuous variable between the specified interval via its `loc` and `scale` arguements. This distribution is constant between `loc` and `loc + scale`.
```python
n = 10000
start = 10
width = 20

data_uniform = uniform.rvs(size = n, loc = start, scale = width)
```

##Gamma Distributive Function
In probability theory and statistics, the gamma distribution is a two-parameter family of continuous probability distributions. The exponential distribution, Erlang distribution, and chi-square distribution are special cases of the gamma distribution. There are two different parameterizations in common use:
1. With a shape parameter k and a scale parameter θ.
2. With a shape parameter α = k and an inverse scale parameter β = 1/θ, called a rate parameter.

In each of these forms, both parameters are positive real numbers.

The corresponding probability density function in the shape-rate parametrization is:

image.png



where ${\displaystyle \Gamma (\alpha )}$ is the gamma function. For all positive integers, ${\displaystyle \Gamma (\alpha )=(\alpha -1)!}$.



```python
from scipy.stats import gamma
data_gamma = gamma.rvs(a = 5, size = 10000)
```

```python
ax = sns.distplot(data_gamma,
                   bins = 30,
                   rug = True,
                   kde = True,
                   color = "blue",
                   hist_kws = {"linewidth":15, "alpha":1})
ax.set(xlabel = "uniform distribution", ylabel = "frequency")
plt.show()
```

##Exponential Distributive Function
In probability theory and statistics, the exponential distribution is the probability distribution of the time between events in a Poisson point process, i.e., a process in which events occur continuously and independently at a constant average rate. It is a particular case of the gamma distribution.
 
image.png


Here λ > 0 is the parameter of the distribution, often called the rate parameter. The distribution is supported on the interval [0, ∞). If a random variable X has this distribution, we write X ~ Exp(λ).

The exponential distribution exhibits infinite divisibility.

```python
from scipy.stats import expon
data_expon = expon.rvs(scale = 1, loc = 0, size = 10000)
```

```python
ax = sns.distplot(data_expon,
                   bins = 100,
                   rug = True,
                   kde = True,
                   color = "blue",
                   hist_kws = {"linewidth":15, "alpha":1})
ax.set(xlabel = "uniform distribution", ylabel = "frequency")
plt.show()
```
